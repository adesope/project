# convert columns to categories

```python
acc['Accident_Severity'] = acc['Accident_Severity'].astype('category')
acc['Day_of_Week'] = acc['Day_of_Week'].astype('category')
acc['Pedestrian_Crossing-Physical_Facilities'] = acc['Pedestrian_Crossing-Physical_Facilities'].astype('category')
acc['Road_Type'] = acc['Road_Type'].astype('category')
acc['Light_Conditions'] = acc['Light_Conditions'].astype('category')
acc['Weather_Conditions'] = acc['Weather_Conditions'].astype('category')
acc['Road_Surface_Conditions'] = acc['Road_Surface_Conditions'].astype('category')
acc['Special_Conditions_at_Site'] = acc['Special_Conditions_at_Site'].astype('category')
acc['Carriageway_Hazards'] = acc['Carriageway_Hazards'].astype('category')
acc['Urban_or_Rural_Area'] = acc['Urban_or_Rural_Area'].astype('category')


# col_cat = ['Accident_Severity','Day_of_Week','Local_Authority_(District)','Road_Type','Pedestrian_Crossing-Physical_Facilities','Light_Conditions','Weather_Conditions','Road_Surface_Conditions','Special_Conditions_at_Site','Carriageway_Hazards','Urban_or_Rural_Area']
# acc = acc[col_cat]
# #acc = [acc[item].astype('category') for item in cat]

# for item in acc[col_cat]:
#     acc = item.astype('category')
# return acc
```

```python
#acc.dtypes
```

```python
acc['Accident_Severity'].cat.categories = ['fatal','serious','slight']
acc['Day_of_Week'].cat.categories = ['Sunday','Monday','Tuesday','Wednesday','Thursday', 'Friday', 'Saturday']
acc['Road_Type'].cat.categories = ['Roundabout','one-way street','Dual carriage','Single carriage','Slip road','Unknown']
acc['Pedestrian_Crossing-Physical_Facilities'].cat.categories = ['No crossing','Zebra','other crossings(Pelican)','Signal junction','Footbridge/Subway','central refuge']
acc['Light_Conditions'].cat.categories = ['Daylight','Darkness-light lits','Darkness-light unlits','Darkness-no light lits','Darkness-unknown']
acc['Weather_Conditions'].cat.categories = ['Fine-no high winds','Raining-no winds','Snowing-no winds','Fine+high winds','Raining+winds', 'Snowing+winds', 'Fog/mist', 'others', 'Unknown']
acc['Road_Surface_Conditions'].cat.categories = ['Dry','Wet/Damp','Snow', 'Frost/ice', 'Flooded (over 3cm deep)']
acc['Special_Conditions_at_Site'].cat.categories = ['None','Traffic signal-out','Traffic signal-defective','Road sign-Defective or obscured', 'Roadworks', 'Road surface-Defective', 'oil/diesel','Mud']
acc['Carriageway_Hazards'].cat.categories = ['None','Vehicle load or road','other object on road', 'Previous accident', 'Pedestrian in carriage way-Not injured', 'Animal in carriage way (except horse)']
acc['Urban_or_Rural_Area'].cat.categories = ['Urban','Rural','Unallocated']
```

```python
acc['Date'] = pd.to_datetime(acc['Date'], dayfirst = True)
```

```python
#chi_square of independence (Light_Conditions and severity)
tab2 = pd.crosstab(acc['Light_Conditions'],acc['Accident_Severity'])
from scipy.stats import chi2_contingency
chi2_2, p_2, dof_2, expected_2 = chi2_contingency(tab2.values)
```
